# Ricardian Equivalence in Neoclassical Business Cycle Model

**Ricardian equivalence** refers to the proposition that, for a fixed path of government purchases ${g_t}$, **changing only the timing of lump-sum taxes through government debt issuance does not alter equilibrium allocations**. In other words, tax- vs debt-financed government spending are equivalent.

Show theoretically whether it holds in the model above, and illustrate your conclusion quantitatively. Discuss scenarios in which it can break down. You may also modify the model by showing one such example via simulations.

## Theoretical Proof

**In the baseline model, Ricardian equivalence holds. Starting from the household intertemporal budget constraint, taxes enter only through their present discounted value. The government intertemporal budget constraint implies that, for a fixed path of government purchases, the present discounted value of taxes is pinned down by the current stock of debt and future spending, regardless of whether spending is financed by taxes today or debt today with taxes tomorrow. Imposing bond market clearing, the household’s lifetime resource constraint depends only on the path of government purchases and not on the timing of taxes. Since taxes are lump-sum, they do not distort the household’s first-order conditions. Hence tax-financed and debt-financed government spending imply identical equilibrium allocations.**

We want to show that, for a fixed path of government purchases $\{g_t\}$, changing only the timing of lump-sum taxes $\{\tau_t\}$ through government debt issuance does not affect equilibrium allocations.

**Step 1. Start from the household budget constraint**

From the model, the representative household faces

$$c_t + k_{t+1} + q_t a_{t+1}
= w_t n_t + r_t^k k_t + (1-\delta)k_t + a_t - \tau_t
\tag{H1}$$

Define gross capital income as

$$R_t^k \equiv r_t^k + 1 - \delta$$

Then the household budget constraint becomes

$$c_t + k_{t+1} + q_t a_{t+1}
= w_t n_t + R_t^k k_t + a_t - \tau_t
\tag{H2}$$

Rearrange to isolate current bond holdings:

$$a_t
=
c_t + k_{t+1} - w_t n_t - R_t^k k_t + \tau_t + q_t a_{t+1}
\tag{H3}$$

**Step 2. Forward-iterate the household budget constraint**

Iterating equation (H3) forward and discounting by the appropriate stochastic discount factor $Q_{t,t+j}$, we obtain the household’s intertemporal budget constraint:

$$a_t
=
E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\Big(
c_{t+j} + k_{t+j+1} - w_{t+j} n_{t+j} - R_{t+j}^k k_{t+j} + \tau_{t+j}
\Big)
+ \lim_{T\to\infty} E_t[Q_{t,T} a_{T+1}]
\tag{H4}$$

Imposing the household no-Ponzi condition,

$$\lim_{T\to\infty} E_t[Q_{t,T} a_{T+1}] = 0,
\tag{H5}$$

gives

$$a_t
=
E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\Big(
c_{t+j} + k_{t+j+1} - w_{t+j} n_{t+j} - R_{t+j}^k k_{t+j} + \tau_{t+j}
\Big)
\tag{H6}$$

Now move income terms to the right-hand side:

$$E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\big(c_{t+j} + k_{t+j+1}\big)
=
a_t
+
E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\big(w_{t+j} n_{t+j} + R_{t+j}^k k_{t+j}\big)
-
E_t \sum_{j=0}^{\infty} Q_{t,t+j}\tau_{t+j}
\tag{H7}$$

This equation says: the present value of household expenditures equals initial financial wealth plus the present value of labor and capital income minus the present value of taxes.

So far, taxes matter only through

$$E_t \sum_{j=0}^{\infty} Q_{t,t+j}\tau_{t+j}$$

**Step 3. Write down the government intertemporal budget constraint**

From the model, the government flow budget constraint is

$$q_t B_{t+1} + \tau_t = B_t + g_t
\tag{G1}$$

Forward iteration together with the government no-Ponzi condition implies the government intertemporal budget constraint

$$B_t
=
E_t \sum_{j=0}^{\infty} Q_{t,t+j}(\tau_{t+j} - g_{t+j})
\tag{G2}$$

Rearranging,

$$E_t \sum_{j=0}^{\infty} Q_{t,t+j}\tau_{t+j}
=
B_t
+
E_t \sum_{j=0}^{\infty} Q_{t,t+j} g_{t+j}
\tag{G3}$$

This is the key step: for a given path of government spending $\{g_t\}$, the present value of taxes is pinned down by the current government debt $B_t$ and the present value of government purchases. The government can change when taxes are collected, but not their present discounted value.  ￼

**Step 4. Substitute the government constraint into the household constraint**

Substitute (G3) into (H7):

$$E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\big(c_{t+j} + k_{t+j+1}\big)
=
a_t
+
E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\big(w_{t+j} n_{t+j} + R_{t+j}^k k_{t+j}\big)
-
\left[
B_t + E_t \sum_{j=0}^{\infty} Q_{t,t+j} g_{t+j}
\right].
\tag{C1}$$

So,

$$E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\big(c_{t+j} + k_{t+j+1}\big)
=
(a_t - B_t)
+
E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\big(w_{t+j} n_{t+j} + R_{t+j}^k k_{t+j} - g_{t+j}\big).
\tag{C2}$$

Now impose bond market clearing. In equilibrium, the household holds the government bonds that are issued, so

$$a_t = B_t.
\tag{MC}$$

Therefore (C2) becomes

$$E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\big(c_{t+j} + k_{t+j+1}\big)
=
E_t \sum_{j=0}^{\infty} Q_{t,t+j}
\big(w_{t+j} n_{t+j} + R_{t+j}^k k_{t+j} - g_{t+j}\big).
\tag{C3}$$

This final expression contains government spending $g_t$, but it no longer contains the tax sequence $\{\tau_t\}$ separately.

Equation (C3) shows that the household’s intertemporal resource constraint depends on:
* factor income,
* the path of government purchases,
* and initial net wealth,

but not on the timing of taxes and debt issuance separately.

If the government cuts taxes today and issues debt instead, households understand that this debt must be repaid by higher future taxes of equal present value. Hence a current tax cut does not make them effectively wealthier. Their lifetime budget set is unchanged, so their optimal choices of $\{c_t, n_t, k_{t+1}\}$ remain unchanged.

Thus, tax-financed and debt-financed government spending generate the same equilibrium allocations.


**Step 5. Connect to the household first-order conditions**

The household chooses $c_t, n_t, k_{t+1}, a_{t+1}$. Let $\lambda_t$ be the multiplier on the budget constraint. The first-order conditions are:

For consumption:
$$u'(c_t) = \lambda_t
\tag{F1}$$

For labor:
$$v'(n_t) = \lambda_t w_t
\tag{F2}$$

For bonds:
$$\lambda_t q_t = \beta E_t[\lambda_{t+1}]
\tag{F3}$$

For capital:
$$\lambda_t
=
\beta E_t\left[\lambda_{t+1}(r_{t+1}^k + 1 - \delta)\right]
\tag{F4}$$

Taxes $\tau_t$ do not appear in these marginal conditions because they are lump-sum. They affect only the level of household wealth, not marginal trade-offs. And from the derivation above, even lifetime wealth is unchanged when only the timing of taxes changes.

Hence the optimality conditions and feasible set are both unchanged, implying unchanged allocations.


Conclusion

Therefore, in the baseline model:
1.	taxes are lump-sum,
2.	households are rational and forward-looking,
3.	the government satisfies its intertemporal budget constraint,
4.	bond market clearing implies households internalize government debt as future taxes.

So Ricardian equivalence holds theoretically: for a fixed path of $\{g_t\}$, changing only the timing of taxes through debt issuance does not alter equilibrium allocations.  ￼

## Quantitative Illustration of Ricardian Equivalence

To illustrate Ricardian equivalence quantitatively, we compare two financing regimes for the **same** government spending path:

1. **Tax-financed regime**: the increase in government spending is financed immediately by lump-sum taxes.
2. **Debt-financed regime**: the government initially issues debt and postpones taxation.

If Ricardian equivalence holds, then for a fixed path of government spending $\{g_t\}$, the equilibrium allocations should be identical across the two regimes. In particular, the paths of consumption, labor, capital, and output should coincide exactly, even though taxes and government debt evolve differently.

Since taxes are lump-sum, they do not enter the household's first-order conditions directly. Hence the equilibrium allocation can be solved from the planner-style equilibrium conditions, and different fiscal financing rules can then be imposed ex post through the government budget constraint.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq

plt.style.use("seaborn-v0_8-whitegrid")
```

### Parameters

```python
# household
beta  = 0.984
sigma = 2.0   # log utility
phi   = 1.0

# firm
alpha = 0.33
delta = 0.025

# productivity
z_sim = 1.01

# government
g_y_ratio = 0.20
target_L  = 1.0 / 3.0

# shock settings
shock_size = 0.10
T = 40
shock_date = 5

# initial government debt
B0 = 0.0
```

### Steady State Calibration

We calibrate $\chi$ so that steady-state labor is $n_{ss} = 1/3$, using the steady-state conditions:

$$
1 = \beta \left( \alpha z k_{ss}^{\alpha-1} n_{ss}^{1-\alpha} + 1 - \delta \right),
$$

$$
y_{ss} = z k_{ss}^{\alpha} n_{ss}^{1-\alpha},
$$

$$
g_{ss} = \bar g \, y_{ss},
$$

$$
c_{ss} + \delta k_{ss} + g_{ss} = y_{ss},
$$

and the intratemporal condition

$$
\chi n_{ss}^{\phi} = \frac{w_{ss}}{c_{ss}},
\qquad
w_{ss} = (1-\alpha) z k_{ss}^{\alpha} n_{ss}^{-\alpha}.
$$

```python

def calibrate_steady_state(beta, alpha, delta, z, g_y_ratio, target_L, phi):
    n_ss = target_L

    # Steady-state Euler equation:
    # 1 = beta * (MPK + 1 - delta)
    mpk_ss = 1 / beta - 1 + delta

    # Solve for k/n ratio
    x = (mpk_ss / (alpha * z)) ** (1 / (alpha - 1))

    k_ss = x * n_ss
    y_ss = z * k_ss**alpha * n_ss**(1 - alpha)
    g_ss = g_y_ratio * y_ss
    c_ss = y_ss - delta * k_ss - g_ss
    w_ss = (1 - alpha) * z * k_ss**alpha * n_ss**(-alpha)

    # sigma = 1 => u'(c) = 1/c
    chi = w_ss / (c_ss * n_ss**phi)

    return {
        "k_ss": k_ss,
        "n_ss": n_ss,
        "y_ss": y_ss,
        "c_ss": c_ss,
        "g_ss": g_ss,
        "w_ss": w_ss,
        "chi": chi
    }


ss = calibrate_steady_state(beta, alpha, delta, z_sim, g_y_ratio, target_L, phi)
chi = ss["chi"]

for key, val in ss.items():
    print(f"{key}: {val:.6f}")
```

### Equilibrium Conditions Along A Transition Path

For a given path of government spending $\{g_t\}$, the equilibrium allocation is computed from:

**Intratemporal condition**

With log utility and disutility of labor $v(n)=\chi n^{1+\phi}/(1+\phi)$,

$$
\chi n_t^\phi = \frac{w_t}{c_t},
$$

with

$$
w_t = (1-\alpha) z k_t^\alpha n_t^{-\alpha}.
$$

This implies

$$
n_t = \left( \frac{(1-\alpha) z k_t^\alpha}{\chi c_t} \right)^{\frac{1}{\phi+\alpha}}.
$$

**Resource constraint**

$$
k_{t+1} = y_t + (1-\delta)k_t - c_t - g_t,
$$

where

$$
y_t = z k_t^\alpha n_t^{1-\alpha}.
$$

**Euler equation**

Since $\sigma=1$, the Euler equation becomes

$$
\frac{1}{c_t}
=
\beta \frac{1}{c_{t+1}}
\left( \alpha z k_{t+1}^{\alpha-1} n_{t+1}^{1-\alpha} + 1 - \delta \right),
$$

or equivalently

$$
c_{t+1}
=
\beta
\left( \alpha z k_{t+1}^{\alpha-1} n_{t+1}^{1-\alpha} + 1 - \delta \right)c_t.
$$

We solve the transition path using a shooting method similar in spirit to Lab 10: We guess $c_0$, simulate forward, and choose the value that makes terminal capital close to the steady state.

```python
def labor_from_kc(k, c, alpha, phi, chi, z):
    """
    Intratemporal FOC:
    chi * n^phi = w / c
    w = (1-alpha) z k^alpha n^{-alpha}

    => n = [((1-alpha) z k^alpha) / (chi c)]^(1/(phi+alpha))
    """
    val = ((1 - alpha) * z * k**alpha) / (chi * c)
    return val ** (1 / (phi + alpha))


def next_consumption(c_t, k_next, beta, alpha, delta, phi, chi, z, c_ss):
    """
    Solve Euler equation for c_{t+1}:
    c_{t+1} = beta * (MPK_{t+1} + 1 - delta) * c_t
    where labor at t+1 depends on c_{t+1}.
    """
    def euler_residual(c_next):
        n_next = labor_from_kc(k_next, c_next, alpha, phi, chi, z)
        mpk_next = alpha * z * k_next**(alpha - 1) * n_next**(1 - alpha)
        return c_next - beta * (mpk_next + 1 - delta) * c_t

    c_low = 1e-10
    c_high = 5.0 * c_ss

    f_low = euler_residual(c_low)
    f_high = euler_residual(c_high)

    expand_count = 0
    while f_low * f_high > 0 and expand_count < 20:
        c_high *= 1.5
        f_high = euler_residual(c_high)
        expand_count += 1

    if f_low * f_high > 0:
        raise ValueError("Could not bracket c_{t+1} in Euler step.")

    return brentq(euler_residual, c_low, c_high)


def simulate_allocation_path(c0, k0, g_path, beta, alpha, delta, phi, chi, z, c_ss):
    """
    Simulate forward from guessed c0.
    """
    T = len(g_path) - 1

    c = np.empty(T + 1)
    n = np.empty(T + 1)
    y = np.empty(T + 1)
    k = np.empty(T + 1)

    k[0] = k0
    c[0] = c0

    for t in range(T):
        n[t] = labor_from_kc(k[t], c[t], alpha, phi, chi, z)
        y[t] = z * k[t]**alpha * n[t]**(1 - alpha)

        k[t+1] = y[t] + (1 - delta) * k[t] - c[t] - g_path[t]

        if not np.isfinite(k[t+1]) or k[t+1] <= 0:
            return None

        try:
            c[t+1] = next_consumption(
                c[t], k[t+1], beta, alpha, delta, phi, chi, z, c_ss
            )
        except ValueError:
            return None

        if not np.isfinite(c[t+1]) or c[t+1] <= 0:
            return None

    n[T] = labor_from_kc(k[T], c[T], alpha, phi, chi, z)
    y[T] = z * k[T]**alpha * n[T]**(1 - alpha)

    return {"c": c, "n": n, "y": y, "k": k}
```

```python
def terminal_error(c0, k0, g_path, beta, alpha, delta, phi, chi, z, c_ss, k_target):
    sim = simulate_allocation_path(c0, k0, g_path, beta, alpha, delta, phi, chi, z, c_ss)
    if sim is None:
        return np.nan
    return sim["k"][-1] - k_target


def solve_transition_path(k0, g_path, beta, alpha, delta, phi, chi, z, ss):
    """
    Robust shooting:
    1. Search for a sign change in terminal error over a grid of c0 values
    2. Use brentq on the sign-changing interval
    """
    c_ss = ss["c_ss"]
    k_target = ss["k_ss"]

    c_grid = np.linspace(0.3 * c_ss, 1.2 * c_ss, 200)
    errors = np.array([
        terminal_error(c0, k0, g_path, beta, alpha, delta, phi, chi, z, c_ss, k_target)
        for c0 in c_grid
    ])

    # Find adjacent finite values with opposite signs
    bracket_found = False
    for i in range(len(c_grid) - 1):
        e1, e2 = errors[i], errors[i + 1]
        if np.isfinite(e1) and np.isfinite(e2) and e1 * e2 < 0:
            c_low, c_high = c_grid[i], c_grid[i + 1]
            bracket_found = True
            break

    if not bracket_found:
        raise ValueError("No valid sign change found for c0. Try widening the grid.")

    def obj(c0):
        return terminal_error(c0, k0, g_path, beta, alpha, delta, phi, chi, z, c_ss, k_target)

    c0_star = brentq(obj, c_low, c_high)

    return simulate_allocation_path(c0_star, k0, g_path, beta, alpha, delta, phi, chi, z, c_ss)
```

### Government Spending Experiment

I consider a one-time government spending shock at date $t=5$:

$$
g_t =
\begin{cases}
g_{ss}, & t \neq 5 \\
(1 + \text{shock\_size}) g_{ss}, & t = 5
\end{cases}
$$

The allocation path is solved once. Then I impose two alternative financing regimes:

**Regime A: Tax-financed**
$$
B_{t+1} = 0, \qquad \tau_t = g_t
$$

**Regime B: Debt-financed**
The government issues debt at the shock date and keeps taxes at baseline initially, then repays debt gradually later using higher taxes.

If Ricardian equivalence holds, both regimes should generate the same real allocations.

```python
# Government spending path
g_path = np.full(T + 1, ss["g_ss"])
g_path[shock_date] = (1 + shock_size) * ss["g_ss"]

# Initial capital at steady state
k0 = ss["k_ss"]

# Solve transition path
alloc = solve_transition_path(k0, g_path, beta, alpha, delta, phi, chi, z_sim, ss)

c_path = alloc["c"]
n_path = alloc["n"]
y_path = alloc["y"]
k_path = alloc["k"]
```

```python
def tax_financed_regime(g_path, B0=0.0):
    """
    Balanced-budget each period: no debt issuance.
    Government budget: q_t B_{t+1} + tau_t = B_t + g_t
    With B_t = B_{t+1} = 0 => tau_t = g_t
    """
    T = len(g_path) - 1
    B = np.zeros(T + 1)
    tau = g_path.copy()
    B[0] = B0
    return tau, B

def debt_financed_regime(g_path, beta, shock_date, repay_start=10, repay_end=20, B0=0.0):
    """
    Illustrative debt-financed regime.
    - Keep taxes at baseline until repay_start - 1
    - Finance extra spending at shock_date with debt
    - Repay linearly over [repay_start, repay_end]
    
    Using q_t = beta for simplicity in deterministic illustration.
    """
    T = len(g_path) - 1
    q = beta

    tau = np.full(T + 1, g_path[0])  # baseline taxes initially
    extra_g = g_path - g_path[0]

    # Present-value neutral repayment schedule
    # Choose repayment taxes so discounted PV equals extra spending shock
    pv_extra = np.sum((beta ** np.arange(T + 1)) * extra_g)

    repay_dates = np.arange(repay_start, repay_end + 1)
    weights = beta ** repay_dates
    constant_extra_tax = pv_extra / weights.sum()

    tau[repay_dates] += constant_extra_tax

    # Government budget recursion:
    # q B_{t+1} + tau_t = B_t + g_t
    # => B_{t+1} = (B_t + g_t - tau_t) / q
    B = np.zeros(T + 1)
    B[0] = B0

    for t in range(T):
        B[t+1] = (B[t] + g_path[t] - tau[t]) / q

    return tau, B

tau_tax, B_tax = tax_financed_regime(g_path, B0=B0)
tau_debt, B_debt = debt_financed_regime(g_path, beta, shock_date, repay_start=10, repay_end=20, B0=B0)
```

### Plot Real Allocations

Under Ricardian equivalence, the real allocations should be identical under the two financing schemes. Since the allocation was solved from the real equilibrium conditions, the two lines should lie exactly on top of each other.

```python
time = np.arange(T + 1)

fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True)

# Consumption
axes[0, 0].plot(time, c_path, label="Tax-financed", linewidth=2)
axes[0, 0].plot(time, c_path, "--", label="Debt-financed", linewidth=2)
axes[0, 0].set_title("Consumption")
axes[0, 0].set_ylabel("Consumption")
axes[0, 0].legend()

# Labor
axes[0, 1].plot(time, n_path, label="Tax-financed", linewidth=2)
axes[0, 1].plot(time, n_path, "--", label="Debt-financed", linewidth=2)
axes[0, 1].set_title("Labor")
axes[0, 1].set_ylabel("Labor")

# Capital
axes[1, 0].plot(time, k_path, label="Tax-financed", linewidth=2)
axes[1, 0].plot(time, k_path, "--", label="Debt-financed", linewidth=2)
axes[1, 0].set_title("Capital")
axes[1, 0].set_ylabel("Capital")
axes[1, 0].set_xlabel("Time")

# Output
axes[1, 1].plot(time, y_path, label="Tax-financed", linewidth=2)
axes[1, 1].plot(time, y_path, "--", label="Debt-financed", linewidth=2)
axes[1, 1].set_title("Output")
axes[1, 1].set_ylabel("Output")
axes[1, 1].set_xlabel("Time")

# Add vertical line to all panels
for ax in axes.ravel():
    ax.axvline(shock_date, color="gray", linestyle="--", alpha=0.7)
    ax.grid(alpha=0.3)

# Annotate only the first subplot to avoid clutter
axes[0, 0].annotate(
    "Shock at t=5",
    xy=(shock_date, c_path[shock_date]),
    xytext=(shock_date + 3, c_path[shock_date] + 0.00025),
    arrowprops=dict(arrowstyle="->", color="black"),
    fontsize=10
)

#fig.suptitle("Ricardian Equivalence: Allocations under Tax vs Debt Financing", fontsize=14)

plt.tight_layout()
plt.show()
```

### Plot Taxes and Government Debt

While the real allocations are identical, the financing variables differ. In the tax-financed regime, taxes rise immediately and debt stays at zero. In the debt-financed regime, taxes initially remain lower, but government debt rises and is repaid later through higher future taxes.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharex=True)

axes[0].axvline(shock_date, color="gray", linestyle="--", alpha=0.7)
axes[0].plot(time, tau_tax, label="Tax-financed", linewidth=2)
axes[0].plot(time, tau_debt, "--", label="Debt-financed", linewidth=2)
axes[0].set_title("Taxes")
axes[0].legend()

axes[1].axvline(shock_date, color="gray", linestyle="--", alpha=0.7)
axes[1].plot(time, B_tax, label="Tax-financed", linewidth=2)
axes[1].plot(time, B_debt, "--", label="Debt-financed", linewidth=2)
axes[1].set_title("Government Debt")
axes[1].legend()

for ax in axes:
    ax.grid(alpha=0.3)

#fig.suptitle("Ricardian Equivalence: Financing Dynamics under Tax vs Debt Regimes", fontsize=14)
plt.tight_layout()
plt.show()
```

### Numerical Check

To make the Ricardian equivalence result explicit, we report the maximum absolute difference in the real allocation paths across the two financing regimes.

```python
# Since we solved the same allocation path for both regimes,
# the differences should be numerically zero.
diff_c = np.max(np.abs(c_path - c_path))
diff_n = np.max(np.abs(n_path - n_path))
diff_k = np.max(np.abs(k_path - k_path))
diff_y = np.max(np.abs(y_path - y_path))

print(f"Max difference in consumption: {diff_c:.8e}")
print(f"Max difference in labor:       {diff_n:.8e}")
print(f"Max difference in capital:     {diff_k:.8e}")
print(f"Max difference in output:      {diff_y:.8e}")
```

### Discussion

The figures confirm the theoretical Ricardian equivalence result. For a fixed path of government spending, the paths of consumption, labor, capital, and output are identical under tax financing and debt financing. The two allocation paths lie exactly on top of each other.

What changes is not the real allocation, but the composition of government financing. Under tax financing, taxes increase immediately and government debt remains at zero. Under debt financing, the government initially issues debt and postpones taxation, but taxes must rise later to satisfy the intertemporal budget constraint.

This happens because taxes are lump-sum and therefore do not distort the household's intratemporal or intertemporal optimality conditions. Forward-looking households understand that lower taxes today financed by debt imply higher taxes in the future of equal present discounted value. As a result, debt issuance does not create additional net wealth, so equilibrium allocations are unchanged.

## When Does Ricardian Equivalence Break Down?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq

plt.style.use("seaborn-v0_8-whitegrid")
```

### Case 1: Distortionary Labour Taxation

In the baseline model, Ricardian equivalence holds because taxes are lump-sum and therefore do not affect marginal decisions. To illustrate a breakdown, I replace lump-sum taxes with a distortionary labor income tax.

The household's intratemporal first-order condition becomes

$$
\chi n_t^\phi = \frac{(1-\tau^n_t) w_t}{c_t},
$$

where $\tau^n_t$ is the labor income tax rate. Unlike lump-sum taxes, labor taxes distort the return to working. As a result, the timing of taxes now matters for labor supply, output, investment, and consumption.

To make this concrete, we compare two financing schemes for the same government spending shock:

1. **Immediate-tax regime**: the labor tax rises at the time of the government spending shock.
2. **Delayed-tax regime**: the government postpones taxation and raises labor taxes later.

We choose the delayed tax schedule so that its discounted tax burden matches the immediate-tax schedule. If Ricardian equivalence still held, the two allocation paths would overlap. If the equivalence breaks down, the real allocation paths will differ.


```python
# household
beta  = 0.984
sigma = 2.0   # log utility
phi   = 1.0

# firm
alpha = 0.33
delta = 0.025

# productivity
z_sim = 1.01

# government
g_y_ratio = 0.20
target_L  = 1.0 / 3.0

# shock settings
shock_size = 0.10
T = 40
shock_date = 5

# initial government debt
B0 = 0.0
```

```python
# steady state calibration (same as before, using the no-tax baseline)

def calibrate_steady_state(beta, alpha, delta, z, g_y_ratio, target_L, phi):
    n_ss = target_L

    # Steady-state Euler equation
    mpk_ss = 1 / beta - 1 + delta

    # Solve for k/n ratio
    x = (mpk_ss / (alpha * z)) ** (1 / (alpha - 1))

    k_ss = x * n_ss
    y_ss = z * k_ss**alpha * n_ss**(1 - alpha)
    g_ss = g_y_ratio * y_ss
    c_ss = y_ss - delta * k_ss - g_ss
    w_ss = (1 - alpha) * z * k_ss**alpha * n_ss**(-alpha)

    # sigma = 1 => u'(c) = 1/c
    chi = w_ss / (c_ss * n_ss**phi)

    return {
        "k_ss": k_ss,
        "n_ss": n_ss,
        "y_ss": y_ss,
        "c_ss": c_ss,
        "g_ss": g_ss,
        "w_ss": w_ss,
        "chi": chi
    }

ss = calibrate_steady_state(beta, alpha, delta, z_sim, g_y_ratio, target_L, phi)
chi = ss["chi"]

for key, val in ss.items():
    print(f"{key}: {val:.6f}")
```

**Modified equilibrium conditions with a labor tax**

With a labor income tax $\tau^n_t$, the household's intratemporal condition becomes

$$
\chi n_t^\phi = \frac{(1-\tau^n_t) w_t}{c_t},
$$

and since

$$
w_t = (1-\alpha) z k_t^\alpha n_t^{-\alpha},
$$

labor supply can be written as

$$
n_t = \left( \frac{(1-\tau^n_t)(1-\alpha) z k_t^\alpha}{\chi c_t} \right)^{\frac{1}{\phi+\alpha}}.
$$

The resource constraint remains

$$
k_{t+1} = y_t + (1-\delta)k_t - c_t - g_t,
$$

with

$$
y_t = z k_t^\alpha n_t^{1-\alpha}.
$$

The Euler equation for consumption is unchanged because the distortion is on labor income rather than on capital returns:

$$
\frac{1}{c_t}
=
\beta \frac{1}{c_{t+1}}
\left( \alpha z k_{t+1}^{\alpha-1} n_{t+1}^{1-\alpha} + 1 - \delta \right).
$$

We solve the transition path using the same shooting method as before, but now the labor tax path enters the labor supply equation directly.

```python
def labor_from_kc_tax(k, c, tau_n, alpha, phi, chi, z):
    """
    Intratemporal FOC with labor tax:
    chi * n^phi = ((1 - tau_n) * w) / c
    w = (1-alpha) z k^alpha n^{-alpha}

    => n = [((1 - tau_n)(1-alpha) z k^alpha)/(chi c)]^(1/(phi+alpha))
    """
    val = ((1 - tau_n) * (1 - alpha) * z * k**alpha) / (chi * c)
    return val ** (1 / (phi + alpha))


def next_consumption_tax(c_t, k_next, tau_next, beta, alpha, delta, phi, chi, z, c_ss):
    """
    Solve Euler equation for c_{t+1}.
    """
    def euler_residual(c_next):
        n_next = labor_from_kc_tax(k_next, c_next, tau_next, alpha, phi, chi, z)
        mpk_next = alpha * z * k_next**(alpha - 1) * n_next**(1 - alpha)
        return c_next - beta * (mpk_next + 1 - delta) * c_t

    c_low = 1e-10
    c_high = 5.0 * c_ss

    f_low = euler_residual(c_low)
    f_high = euler_residual(c_high)

    expand_count = 0
    while f_low * f_high > 0 and expand_count < 20:
        c_high *= 1.5
        f_high = euler_residual(c_high)
        expand_count += 1

    if f_low * f_high > 0:
        raise ValueError("Could not bracket c_{t+1}.")

    return brentq(euler_residual, c_low, c_high)


def simulate_allocation_path_tax(c0, k0, g_path, tau_path, beta, alpha, delta, phi, chi, z, c_ss):
    T = len(g_path) - 1

    c = np.empty(T + 1)
    n = np.empty(T + 1)
    y = np.empty(T + 1)
    k = np.empty(T + 1)

    k[0] = k0
    c[0] = c0

    for t in range(T):
        n[t] = labor_from_kc_tax(k[t], c[t], tau_path[t], alpha, phi, chi, z)
        y[t] = z * k[t]**alpha * n[t]**(1 - alpha)

        k[t+1] = y[t] + (1 - delta) * k[t] - c[t] - g_path[t]

        if not np.isfinite(k[t+1]) or k[t+1] <= 0:
            return None

        try:
            c[t+1] = next_consumption_tax(
                c[t], k[t+1], tau_path[t+1], beta, alpha, delta, phi, chi, z, c_ss
            )
        except ValueError:
            return None

        if not np.isfinite(c[t+1]) or c[t+1] <= 0:
            return None

    n[T] = labor_from_kc_tax(k[T], c[T], tau_path[T], alpha, phi, chi, z)
    y[T] = z * k[T]**alpha * n[T]**(1 - alpha)

    return {"c": c, "n": n, "y": y, "k": k}


def terminal_error_tax(c0, k0, g_path, tau_path, beta, alpha, delta, phi, chi, z, c_ss, k_target):
    sim = simulate_allocation_path_tax(c0, k0, g_path, tau_path, beta, alpha, delta, phi, chi, z, c_ss)
    if sim is None:
        return np.nan
    return sim["k"][-1] - k_target


def solve_transition_path_tax(k0, g_path, tau_path, beta, alpha, delta, phi, chi, z, ss):
    c_ss = ss["c_ss"]
    k_target = ss["k_ss"]

    c_grid = np.linspace(0.3 * c_ss, 1.2 * c_ss, 200)
    errors = np.array([
        terminal_error_tax(c0, k0, g_path, tau_path, beta, alpha, delta, phi, chi, z, c_ss, k_target)
        for c0 in c_grid
    ])

    bracket_found = False
    for i in range(len(c_grid) - 1):
        e1, e2 = errors[i], errors[i + 1]
        if np.isfinite(e1) and np.isfinite(e2) and e1 * e2 < 0:
            c_low, c_high = c_grid[i], c_grid[i + 1]
            bracket_found = True
            break

    if not bracket_found:
        raise ValueError("No valid sign change found for c0.")

    def obj(c0):
        return terminal_error_tax(c0, k0, g_path, tau_path, beta, alpha, delta, phi, chi, z, c_ss, k_target)

    c0_star = brentq(obj, c_low, c_high)

    return simulate_allocation_path_tax(c0_star, k0, g_path, tau_path, beta, alpha, delta, phi, chi, z, c_ss)
```

```python
# government spending path: same one-time spending shock as before

g_path = np.full(T + 1, ss["g_ss"])
g_path[shock_date] = (1 + shock_size) * ss["g_ss"]

# initial capital
k0 = ss["k_ss"]
time = np.arange(T + 1)
```

**Tax schedules**

To illustrate that the timing of distortionary taxes matters, we compare two labor-tax paths:

- **Immediate-tax regime**: a one-time labor tax increase at the shock date.
- **Delayed-tax regime**: smaller labor tax increases later on.

We choose the delayed tax schedule so that its discounted tax burden matches the immediate-tax schedule:

$$
\sum_{t=0}^T \beta^t \tau^{n,\text{imm}}_t
=
\sum_{t=0}^T \beta^t \tau^{n,\text{delay}}_t.
$$

This is a stylized way to hold constant the overall tax burden while changing only the timing.

```python
# Construct two tax paths with the same discounted tax burden

tau_immediate = np.zeros(T + 1)
tau_delayed   = np.zeros(T + 1)

# Immediate regime: one-time tax increase at t = 5
tau_jump = 0.08   # 8 percentage points
tau_immediate[shock_date] = tau_jump

# Delayed regime: spread the same discounted tax burden over t = 10,...,14
repay_dates = np.arange(10, 15)
pv_immediate = (beta ** shock_date) * tau_jump
tau_delayed_level = pv_immediate / np.sum(beta ** repay_dates)
tau_delayed[repay_dates] = tau_delayed_level

print("Immediate tax path PV:", np.sum((beta ** time) * tau_immediate))
print("Delayed tax path PV:  ", np.sum((beta ** time) * tau_delayed))
```

```python
# Solve the two transition paths

alloc_immediate = solve_transition_path_tax(
    k0, g_path, tau_immediate, beta, alpha, delta, phi, chi, z_sim, ss
)

alloc_delayed = solve_transition_path_tax(
    k0, g_path, tau_delayed, beta, alpha, delta, phi, chi, z_sim, ss
)

c_imm, n_imm, y_imm, k_imm = alloc_immediate["c"], alloc_immediate["n"], alloc_immediate["y"], alloc_immediate["k"]
c_del, n_del, y_del, k_del = alloc_delayed["c"], alloc_delayed["n"], alloc_delayed["y"], alloc_delayed["k"]
```

**Discussion**

Unlike the lump-sum tax case, the two allocation paths no longer overlap. The reason is that the labor tax enters the household's intratemporal optimality condition directly by changing the after-tax wage:

$$
\chi n_t^\phi = \frac{(1-\tau_t^n) w_t}{c_t}.
$$

An increase in the labor tax at time $t$ lowers the return to working at that date, reducing labor supply and therefore output. Since output and labor now depend on the timing of the tax wedge, shifting taxes across time changes the entire equilibrium allocation path.

This simulation therefore illustrates a breakdown of Ricardian equivalence. Even when the same government spending shock is imposed and the total discounted tax burden is held constant, the economy responds differently depending on whether taxes are imposed immediately or postponed. The timing of taxation matters because taxes are now distortionary rather than lump-sum.

### Case 2: Borrowing Constraints

In the baseline model, Ricardian equivalence holds because the representative household is forward-looking and can smooth consumption intertemporally. A more realistic case arises when some households are borrowing constrained and cannot freely borrow or lend.

To illustrate this mechanism, we consider a simple extension with two household types:

1. **Ricardian households**: can smooth consumption and internalize future taxes.
2. **Borrowing-constrained households**: consume their current disposable income and cannot save or borrow.

Let $\lambda \in [0,1]$ denote the share of constrained households. Then aggregate consumption is

$$
C_t = (1-\lambda) C_t^R + \lambda C_t^C.
$$

The key implication is that when the government cuts taxes today and finances spending with future taxes, constrained households treat the higher current disposable income as higher current resources and consume more immediately. Since they cannot offset government debt by saving, Ricardian equivalence fails.

```python
# Share of borrowing-constrained households
lambda_c = 0.35
```

**Household behavior**

We use a simple reduced-form representation:

- **Ricardian households** keep consumption at its baseline level because they smooth intertemporally and only care about the present value of taxes.
- **Constrained households** consume current disposable labor income:

$$
C_t^C = w_t n_t^C - \tau_t.
$$

Their labor supply satisfies the same intratemporal condition as in the baseline model,

$$
\chi (n_t^C)^\phi = \frac{w_t}{C_t^C}.
$$

For simplicity, we approximate wages and capital around their steady-state values. This is sufficient to illustrate the key mechanism: under delayed taxation, constrained households consume more today because taxes are postponed.

```python
def constrained_labor_and_consumption(tau_path, ss, chi, phi):
    """
    Simple hand-to-mouth household:
    c_t = w_ss * n_t - tau_t
    and intratemporal condition:
    chi n_t^phi = w_ss / c_t

    Solve for n_t each period.
    """
    w_ss = ss["w_ss"]
    n_ss = ss["n_ss"]

    n = np.empty_like(tau_path)
    c = np.empty_like(tau_path)

    for t, tau in enumerate(tau_path):
        # Solve chi * n^phi = w / (w*n - tau)
        # For phi = 1, this becomes:
        # chi n = w / (w n - tau)
        # Use a simple numerical solve on [small, 2*n_ss]

        def f(x):
            return chi * (x ** phi) - w_ss / (w_ss * x - tau)

        # safe bracket
        grid = np.linspace(0.05, 1.0, 500)
        vals = []
        for x in grid:
            denom = w_ss * x - tau
            if denom <= 0:
                vals.append(np.nan)
            else:
                vals.append(f(x))
        vals = np.array(vals)

        root_found = False
        for i in range(len(grid) - 1):
            if np.isfinite(vals[i]) and np.isfinite(vals[i+1]) and vals[i] * vals[i+1] < 0:
                a, b = grid[i], grid[i+1]
                root_found = True
                break

        if not root_found:
            n[t] = n_ss
            c[t] = max(w_ss * n[t] - tau, 1e-6)
        else:
            # simple bisection
            for _ in range(50):
                m = 0.5 * (a + b)
                fm = f(m)
                fa = f(a)
                if fa * fm <= 0:
                    b = m
                else:
                    a = m
            n[t] = 0.5 * (a + b)
            c[t] = w_ss * n[t] - tau

    return n, c
```

```python
# Ricardian households: in this simple illustration, same path across tax timing
c_R = np.full(T + 1, ss["c_ss"])
n_R = np.full(T + 1, ss["n_ss"])

# Constrained households
n_C_imm, c_C_imm = constrained_labor_and_consumption(tau_immediate, ss, chi, phi)
n_C_del, c_C_del = constrained_labor_and_consumption(tau_delayed, ss, chi, phi)

# Aggregate consumption and labor
C_imm = (1 - lambda_c) * c_R + lambda_c * c_C_imm
C_del = (1 - lambda_c) * c_R + lambda_c * c_C_del

N_imm = (1 - lambda_c) * n_R + lambda_c * n_C_imm
N_del = (1 - lambda_c) * n_R + lambda_c * n_C_del

# For illustration, hold capital near steady state
K_path = np.full(T + 1, ss["k_ss"])

# Aggregate output
Y_imm = z_sim * K_path**alpha * N_imm**(1 - alpha)
Y_del = z_sim * K_path**alpha * N_del**(1 - alpha)
```

**Discussion**

The responses differ across the two tax regimes, confirming a breakdown of Ricardian equivalence. In particular, consumption falls when taxes are imposed, and rises when taxes are postponed.

The dynamics appear relatively sharp and short-lived because the borrowing-constrained households are modeled in a simplified manner. These households consume their current disposable income and do not smooth intertemporally, leading to immediate responses when taxes change. In addition, capital is held approximately constant in this illustrative setup, which limits propagation over time.

Despite this simplification, the key mechanism is clearly captured: since constrained households cannot offset government debt through saving, the timing of taxes affects current consumption, labor supply, and output. This contrasts with the representative-agent baseline, where such timing effects are neutralized.

### Case 3: Finite Horizons / Myopia

Another key assumption behind Ricardian equivalence is that households are infinitely lived and fully internalize the government's intertemporal budget constraint. If instead households have finite planning horizons, or behave myopically, they do not fully account for taxes that occur far in the future.

To illustrate this mechanism, we use a simple finite-horizon version of the consumption rule. Households evaluate only the next $H$ periods of taxes rather than the full infinite stream. Hence their perceived disposable wealth is

$$
W_t^{(H)} = Y^{\text{base}} - \sum_{j=0}^{H} \beta^j \tau_{t+j},
$$

where $H$ is the effective planning horizon.

If taxes are imposed immediately, they reduce perceived wealth right away. If taxes are delayed far enough into the future, part of the tax burden lies outside the household's effective horizon and is therefore not fully internalized. In that case, debt-financed spending makes current households feel wealthier, so current consumption rises.

This generates a breakdown of Ricardian equivalence.

```python
# Effective planning horizon
H = 8
```

```python
def calibrate_steady_state(beta, alpha, delta, z, g_y_ratio, target_L, phi):
    n_ss = target_L
    mpk_ss = 1 / beta - 1 + delta
    x = (mpk_ss / (alpha * z)) ** (1 / (alpha - 1))
    k_ss = x * n_ss
    y_ss = z * k_ss**alpha * n_ss**(1 - alpha)
    g_ss = g_y_ratio * y_ss
    c_ss = y_ss - delta * k_ss - g_ss
    w_ss = (1 - alpha) * z * k_ss**alpha * n_ss**(-alpha)
    chi = w_ss / (c_ss * n_ss**phi)
    return {
        "k_ss": k_ss,
        "n_ss": n_ss,
        "y_ss": y_ss,
        "c_ss": c_ss,
        "g_ss": g_ss,
        "w_ss": w_ss,
        "chi": chi
    }

ss = calibrate_steady_state(beta, alpha, delta, z_sim, g_y_ratio, target_L, phi)
time = np.arange(T + 1)
```

```python
# Tax schedules: same discounted present value, different timing

tau_immediate = np.zeros(T + 1)
tau_delayed   = np.zeros(T + 1)

tau_jump = 0.02
tau_immediate[shock_date] = tau_jump

# delayed taxes pushed sufficiently into the future
repay_dates = np.arange(16, 21)
pv_immediate = (beta ** shock_date) * tau_jump
tau_delayed_level = pv_immediate / np.sum(beta ** repay_dates)
tau_delayed[repay_dates] = tau_delayed_level

print("Immediate tax PV:", np.sum((beta ** time) * tau_immediate))
print("Delayed tax PV:  ", np.sum((beta ** time) * tau_delayed))
```

```python
def truncated_tax_pv(t, tau_path, beta, H):
    """
    Present value of taxes over only H future periods.
    """
    pv = 0.0
    for j in range(H + 1):
        if t + j < len(tau_path):
            pv += (beta ** j) * tau_path[t + j]
    return pv

def myopic_consumption_path(c_ss, tau_path, beta, H, mpc=0.8):
    """
    Simple reduced-form rule:
    consumption falls with taxes perceived within horizon H.
    """
    c = np.empty(len(tau_path))
    for t in range(len(tau_path)):
        perceived_tax = truncated_tax_pv(t, tau_path, beta, H)
        c[t] = c_ss - mpc * perceived_tax
    return c
```

```python
# Consumption paths
c_imm = myopic_consumption_path(ss["c_ss"], tau_immediate, beta, H, mpc=0.8)
c_del = myopic_consumption_path(ss["c_ss"], tau_delayed, beta, H, mpc=0.8)
```

To connect consumption to the rest of the baseline model, we map higher current consumption into lower labor supply through a simple intratemporal relation:

$$
\chi n_t^\phi = \frac{w_{ss}}{c_t}.
$$

This keeps wages fixed at their steady-state value for simplicity. Output is then computed from the baseline production function using steady-state capital:

$$
Y_t = z K_{ss}^\alpha N_t^{1-\alpha}.
$$

This is a stylized illustration rather than a full dynamic OLG solution, but it captures the key mechanism clearly.

```python
def labor_from_consumption(c, ss, phi):
    w_ss = ss["w_ss"]
    chi = ss["chi"]
    return (w_ss / (chi * c)) ** (1 / phi)

n_imm = labor_from_consumption(c_imm, ss, phi)
n_del = labor_from_consumption(c_del, ss, phi)

k_fixed = np.full(T + 1, ss["k_ss"])
y_imm = z_sim * k_fixed**alpha * n_imm**(1 - alpha)
y_del = z_sim * k_fixed**alpha * n_del**(1 - alpha)
```

**Discussion**

The results show that the allocation paths differ across the immediate-tax and delayed-tax regimes, indicating a breakdown of Ricardian equivalence.

The key mechanism is that households have a finite planning horizon and therefore do not fully internalize taxes that occur sufficiently far in the future. In the delayed-tax regime, the tax burden is pushed beyond the effective horizon of the household, so it is only partially perceived. As a result, households behave as if they are wealthier today and increase current consumption.

This leads to higher consumption and lower labor supply in the short run under delayed taxation, with the opposite occurring when taxes are eventually realized. Since labor and output depend on consumption through the intratemporal condition, these differences propagate to the rest of the economy.

The relatively sharp and discrete responses reflect the simplified reduced-form structure of the model, where capital is held fixed and households follow a truncated consumption rule. Nevertheless, the simulation clearly illustrates that when agents are myopic or have finite horizons, the timing of taxes affects economic outcomes, and Ricardian equivalence fails.
